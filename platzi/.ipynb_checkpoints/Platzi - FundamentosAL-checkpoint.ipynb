```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

xy_range = (-3, 3)
x = np.arange(xy_range[0], xy_range[1], 0.5)
y_1 = 3*x + 5
y_2 = 2*x + 3

plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)

plt.xlim(xy_range)
plt.ylim(xy_range)

plt.axvline(x=0, color='gray')
plt.axhline(y=0, color='gray')
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

xy_range = (-6, 6)
x = np.arange(xy_range[0], xy_range[1], 0.5)
y_1 = 3*x + 5
y_2 = -x + 3
y_3 = 2*x + 1

plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(xy_range)
plt.ylim(xy_range)

plt.axvline(x=0, color='gray')
plt.axhline(y=0, color='gray')
plt.show()
```
